```python
df = pd.read_csv('main_task.xls', low_memory = False)
df['Cuisine Style'] = df['Cuisine Style'].fillna('Nothing')
list_cuisin = []
for i in df['Cuisine Style']:
    i = i.strip("[]")
    i = i.replace("\'", '')
    i = i.split(", ")
    list_cuisin.append(i)
df['Cuisine Style'] = list_cuisin
df['cuis_len'] = df['Cuisine Style'].apply(lambda x: len(x))
review_dates = []
for i in df['Reviews']:
    result = re.findall(r'\d{2}/\d{2}/\d{4}', i)
    review_dates.append(result)
df['Review difference'] = review_dates
df['# of reviews'] = df['Review difference'].apply(lambda x: len(x))
dif_list = df['Review difference'].apply(lambda x: pd.to_datetime(x[0])-pd.to_datetime(x[1]) if len(x) == 2 else pd.to_datetime(0)-pd.to_datetime(0))
weekday = df['Review difference'].apply(lambda x: pd.to_datetime(x[0]).weekday() if len(x)==2 or len(x)==1 else -1)
df['weekday1st review'] = weekday
df['Review difference'] = dif_list.dt.days
df['Price Range'] = df['Price Range'].fillna(0)

def sign_to_value(x):
    if x == '$$ - $$$':
        return 2
    elif x == '$':
        return 1
    elif x == '$$$$':
        return 3
    else:
        return 0
    
list_of = df['Price Range'].apply(lambda x: sign_to_value(x))
df['Price value'] = list_of

dsa = set()
for i in df['Cuisine Style']:
    for q in i:
        dsa.add(q)
for i in dsa:
    df[i] = 0
for index, row in df.iterrows():
    for i in row['Cuisine Style']:
        df[i][index] = 1
        
df1 = pd.get_dummies(df.City)
for i in df1:
    df[i] = df1[i]
```
