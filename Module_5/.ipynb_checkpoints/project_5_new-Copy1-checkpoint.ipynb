```python
def get_boxplot(column):
    fig, ax = plt.subplots(figsize = (14, 4))
    sns.boxplot(x=column,
                data=train[train.default>0],ax=ax)
    plt.xticks(rotation=45)
    ax.set_title('Boxplot for ' + column)
    plt.show()
```

```python
# Based on the above, we should rework two datasets, first for train and second for test
```

```python
# Firstly we can drop cliend_id column as it does not bring any value
train.drop('client_id', axis=1, inplace=True)
```

```python
train.education.value_counts().plot.barh()
```

```python
# fill na values in education array, the most common value for education is SCH
# the loop that fills nan values with its most suitable values based on the same info
for index, row in train[train.education.isnull()].iterrows():
    a = train['education'][(train.age == row[5]) & (train.sex == row[4]) & (train.car == row[6]) & (train.car_type == row[7]) & (train.decline_app_cnt == row[8]) & 
       (train.good_work == row[9]) & (train.bki_request_cnt == row[11])  & (train.income == row[15])].value_counts() # looks for the same info
    if len(a) != 0: # if value_counts is not empty then it fills with its #1 value
        train['education'][index] = a.index[0]
    else: # else fills ir with the most common value
        train['education'][index] = 'SCH'

for index, row in test[test.education.isnull()].iterrows():
    a = test['education'][(test.age == row[5]) & (test.sex == row[4]) & (test.car == row[6]) & (test.car_type == row[7]) & (test.decline_app_cnt == row[8]) & 
       (test.good_work == row[9]) & (test.bki_request_cnt == row[11])  & (test.income == row[15])].value_counts() # looks for the same info
    if len(a) != 0: # if value_counts is not empty then it fills with its #1 value
        test['education'][index] = a.index[0]
    else: # else fills ir with the most common value
        test['education'][index] = 'SCH'
```

```python
# the remaining app_date should be converted to a int type
bin_cols = ['sex', 'car', 'car_type', 'foreign_passport', 'good_work']
cat_cols = ['education', 'work_address', 'home_address', 'sna','first_time']
num_cols = ['age', 'decline_app_cnt', 'bki_request_cnt', 'income', 'score_bki', 'region_rating']
```

```python
# Converting date to timestamp format
train['app_date'] = pd.to_datetime(train['app_date'])
test['app_date'] = pd.to_datetime(test['app_date'])
```

```python
# Creating new column with a month of a date and its weekday
train['month'] = train.app_date.dt.month
train['weekday'] = train.app_date.dt.dayofweek
test['month'] = train.app_date.dt.month
test['weekday'] = train.app_date.dt.dayofweek
train.drop('app_date', axis=1, inplace=True)
test.drop('app_date', axis=1, inplace=True)
# We should also add the new columns to our lists
cat_cols.append('month')
cat_cols.append('weekday')
```

```python
# num_cols data exploration
a,b,c = 2,3,1
fig = plt.figure(figsize=(14,10))
for col in num_cols:
    plt.subplot(a, b, c)
    plt.xlabel(col)
    sns.distplot(train[col])
    c = c + 1
plt.show()
# Based on output score_bki looks like it is scaled, age is about to be scaled and the rest is not with a huge right tail
```

```python
# let's check how it changes if we take log numbers
a,b,c = 2,3,1 
fig = plt.figure(figsize=(14,10))
for col in num_cols:
        plt.subplot(a, b, c)
        plt.xlabel(col)
        sns.distplot(np.log(train[col]+1)) # since we have 0 in df we simply add 1 to each value
        c = c + 1
plt.show()
```

```python
plt.subplots(figsize=(10, 10))
sns.heatmap(train[num_cols].corr(), square=True,
              annot=True, fmt=".5f", linewidths=0.1, cmap="RdBu")
```

```python
for i in num_cols:
    get_boxplot(i)
```

```python
imp_num = pd.Series(f_classif(train[num_cols], train['default'])[0], index = num_cols)
imp_num.sort_values(inplace = True)
imp_num.plot(kind = 'barh')
```

```python
# There is no too strong coorelation between numerical values, however, 
#there is a noticable thing that income and region_rating has the strongest coorelation among the num_cols
label_encoder = LabelEncoder()

for column in bin_cols:
    train[column] = label_encoder.fit_transform(train[column])
    
for column in bin_cols:
    test[column] = label_encoder.fit_transform(test[column])
```

```python
train['education'] = train['education'].astype('category').cat.reorder_categories(['SCH','GRD','UGR','PGR','ACD']).cat.codes
test['education'] = test['education'].astype('category').cat.reorder_categories(['SCH','GRD','UGR','PGR','ACD']).cat.codes
```

```python
imp_cat = Series(mutual_info_classif(train[bin_cols + cat_cols], train['default'],
                                     discrete_features =True), index = bin_cols + cat_cols)
imp_cat.sort_values(inplace = True)
imp_cat.plot(kind = 'barh')
```

```python
imp_num = Series(f_classif(train[num_cols], train['default'])[0],
                 index=num_cols)
imp_num.sort_values(inplace=True)
imp_num.plot(kind='barh',grid=True)
plt.xlabel('F-value')
plt.show()
```

```python
# Work address and home address has the strongest coorelation 0.7. I decided to remove on of these. Let it be home address\
plt.subplots(figsize=(15, 15))
sns.heatmap(train.corr(), square=True,
              annot=True, fmt=".1f", linewidths=0.1, cmap="RdBu")
```

```python
train.drop('home_address', axis=1, inplace=True)
test.drop('home_address', axis=1, inplace=True)
cat_cols.remove('home_address')
```

```python
###################################################################
##################################################################
###################################################################
train.income = np.log(train.income)
train.age = np.log(train.age)
```

```python
X_cat = OneHotEncoder(sparse = False).fit_transform(train[cat_cols].values)
Y_cat = OneHotEncoder(sparse = False).fit_transform(test[cat_cols].values)
```

```python
X_num = StandardScaler().fit_transform(train[num_cols].values)
Y_num = StandardScaler().fit_transform(test[num_cols].values)
```

```python
X = np.hstack([X_num, train[bin_cols].values, X_cat])
Y = train['default'].values

id_test = test['client_id']
test = np.hstack([Y_num, test[bin_cols].values, Y_cat])
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42, shuffle=True)
```

```python
from sklearn.model_selection import GridSearchCV
model = LogisticRegression(random_state=42, max_iter=5000)
C = np.logspace(0, 5, 10)
param_grid = [
    {'penalty': ['l2'], 
     'solver': ['lbfgs', 'liblinear'],
    'C': C}
#     {'penalty': ['l2'], 
#      'solver': ['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga'], 
#      'class_weight':['none', 'balanced'], 
#      'multi_class': ['auto','ovr'],
#     'C': C},
#     {'penalty': ['none'], 
#      'solver': ['newton-cg', 'lbfgs', 'sag', 'saga'], 
#      'class_weight':['none', 'balanced'], 
#      'multi_class': ['auto','ovr'],
#     'C': C}
]
gridsearch = GridSearchCV(model, param_grid, cv=5, verbose=0)
gridsearch.fit(X_train, y_train)
model = gridsearch.best_estimator_
## печатаем параметры
best_parameters = model.get_params()
best_parameters
```

```python
# Описываем и обучаем модель
model = LogisticRegression(
    C=3.5938136638046276,
    class_weight = None,
    dual = False,
    fit_intercept = True,
    intercept_scaling= 1,
    l1_ratio = None,
    max_iter = 5000,
    multi_class = 'auto',
    n_jobs = None,
    penalty = 'l2',
    random_state = 42,
    solver = 'lbfgs',
    tol = 0.0001,
    verbose = 0,
    warm_start = False
)

model.fit(X_train, y_train)
```

```python
Y_predict = model.predict(X_test)
Y_predict_prob = model.predict_proba(X_test)[:,1]
```

```python
y_pred_test = model.predict(test)
y_pred_prob_test = model.predict_proba(test)[:,1]
```

```python
fpr, tpr, threshold = roc_curve(y_test, Y_predict_prob)
roc_auc = roc_auc_score(y_test, Y_predict_prob)

plt.figure()
plt.plot([0, 1], label='Baseline', linestyle='--')
plt.plot(fpr, tpr, label = 'Regression')
plt.title('Logistic Regression ROC AUC = %0.3f' % roc_auc)
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.legend(loc = 'lower right')
plt.show()
```

```python
main_model = LogisticRegression( 
    C=3.5938136638046276,
    class_weight = None,
    dual = False,
    fit_intercept = True,
    intercept_scaling= 1,
    l1_ratio = None,
    max_iter = 5000,
    multi_class = 'auto',
    n_jobs = None,
    penalty = 'l2',
    random_state = 42,
    solver = 'lbfgs',
    tol = 0.0001,
    verbose = 0,
    warm_start = False
)
main_model.fit(X, Y)

# Предсказываем значения тестовой базы
y_pred_test = main_model.predict(test)
y_pred_prob_test = main_model.predict_proba(test)[:,1]
```

```python
new_sample_submission = pd.DataFrame({'client_id': id_test,
                              'default': y_pred_prob_test})
new_sample_submission.to_csv('submission.csv', index=False)

new_sample_submission.head(10)
```
